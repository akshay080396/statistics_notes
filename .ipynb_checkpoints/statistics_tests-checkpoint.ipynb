Average daily steps

To validate its marketing claims and understand user behavior, the company collected data from 100 users enrolled in three fitness programs ‚Äî Cardio, Strength, and Yoga.

#### 1Ô∏è‚É£ Claim 1: Average Steps per Day (Z-Test / One-Sample t-Test)
‚ÄúOn average, our users walk 8,000 steps per day.‚Äù

H0= mean of population is 8000
H1 = mean not equal to 8000

```python
df.shape
```

```python
n=100
```

```python
X_population=8000
```

```python
x_sample=df['Avg_Steps'].mean()
x_sample
```

```python
x_std_sample=df['Avg_Steps'].std()
x_std_sample
```

```python
z_test=(x_sample-X_population)/(x_std_sample/(math.sqrt(100)))
```

```python
z_test
```

```python
p=2*(1-norm.cdf(abs(z_test)))
```

```python
p
```

##### p = 0.2439 > 0.05 ‚Üí Fail to reject H‚ÇÄ

Meaning: Your data does not show a statistically significant difference from 8000 steps/day.

The company‚Äôs claim holds true given your sample.

### Do users enrolled in the ‚ÄúCardio‚Äù program have a different average resting heart rate than those in the ‚ÄúYoga‚Äù program?

```python
df.head()
```

```python
df_yoga=df[df['Program']=='Yoga']
df_yoga
```

```python
df_cardio=df[df['Program']=='Cardio']
df_cardio
```

```python
yoga_std=df_yoga['Heart_Rate'].std()
```

```python
cardio_mean=df_cardio['Heart_Rate'].mean()
```

```python
cardio_std=df_cardio['Heart_Rate'].std()
```

```python
t_stat, p_value = ttest_ind(df_cardio['Heart_Rate'],df_yoga['Heart_Rate'], equal_var=False)

```

```python
print(f"T-statistic = {t_stat:.2f}")
print(f"P-value = {p_value:.4f}")
```

| Metric         | Interpretation                                             |
| -------------- | ---------------------------------------------------------- |
| **T = -0.17**  | Very small difference (almost no effect)                   |
| **P = 0.8686** | High probability that observed difference is due to chance |
| **Conclusion** | No significant difference between groups                   |


### üè¢ Company Claim:
‚ÄúGender and program choice are independent ‚Äî users choose their programs equally regardless of gender.‚Äù

Goal: Check if there is any association between gender and program enrollment.

```python
df.head()
```

```python
df['Gender'].value_counts()
```

```python
df['Program'].value_counts()
```

```python
from scipy.stats import chi2_contingency
```

```python
cont_table = pd.crosstab(df["Program"], df["Gender"])
print(cont_table)
```

```python
chi2, p, dof, expected = chi2_contingency(cont_table)
print("Chi-square value:", chi2)
print("Degrees of freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies:\n", expected)
```

| Step | Description              | Python Function                      |
| ---- | ------------------------ | ------------------------------------ |
| 1    | Create contingency table | `pd.crosstab()`                      |
| 2    | Perform Chi-square test  | `chi2_contingency()`                 |
| 3    | Check p-value            | `if p < 0.05:`                       |
| 4    | Visualize                | `sns.countplot()` or `sns.heatmap()` |


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=df, x="Program", hue="Gender", palette="Set2")
plt.title("Program Enrollment by Gender")
plt.show()
```

#### Average daily calories burned are the same across all three programs ‚Äî Cardio, Strength, and Yoga.

```python
from scipy.stats import f_oneway
```

```python
cardio = df[df['Program'] == 'Cardio']['Calories_Burned']
cardio
```

```python
cardio = df[df['Program'] == 'Cardio']['Calories_Burned']
strength = df[df['Program'] == 'Strength']['Calories_Burned']
yoga = df[df['Program'] == 'Yoga']['Calories_Burned']
```

```python
F_stat, p_value = f_oneway(cardio, strength, yoga)
print(f"F-statistic = {F_stat:.4f}")
print(f"P-value = {p_value:.4f}")
```
